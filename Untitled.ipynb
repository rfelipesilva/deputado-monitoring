### Getting deputy id

```python
deputies_call = 'https://dadosabertos.camara.leg.br/api/v2/deputados?ordem=ASC&ordenarPor=nome'
```

```python
response = requests.get(deputies_call)
data = response.json()
```

```python
deputy_ids = [_id['id'] for _id in data['dados']]
```

```python
len(deputy_ids)
```

```python
deputy_ids[1]
```

```python
years = [2019, 2020, 2021]

dfs_deputy_to_contact = []

count = 0

for deputy_id in deputy_ids:
    
    count += 1

    print(count, ' - ', deputy_id)

    api = 'https://dadosabertos.camara.leg.br/api/v2/deputados/{}/'.format(deputy_id)

    cost_year, cost_month, cost_type, cost_value, cost_code, deputy_unique = [], [], [], [], [], []

    for year in years:

        method = f'despesas?ano={year}&ordem=ASC&ordenarPor=ano'
        url_call = api + method
        time.sleep(2.5)
        response = requests.get(url_call)

        data = response.json()

        # print(count, ' - ', deputy_id, f'- response:{response.status_code}')
        # print(url_call)

        for costs in data['dados']:
            cost_year.append(costs['ano'])
            cost_month.append(costs['mes'])
            cost_type.append(costs['tipoDespesa'])
            cost_value.append(costs['valorDocumento'])
            cost_code.append(costs['codDocumento'])
            deputy_unique.append(deputy_id)

            df = pd.DataFrame({'year': cost_year, 'month': cost_month, 'cost_type': cost_type, 
                                'value': cost_value, 'cost_code': cost_code, 'deputy_id':deputy_unique})
            # print(df)
    df.to_csv('data/csv/{}.csv'.format(deputy_id), sep='|', encoding='utf-8', index=False)
    dfs_deputy_to_contact.append(df)

costs_df = pd.concat(dfs_deputy_to_contact).drop_duplicates()
costs_df.to_csv('data/csv/costs_full.csv', sep='|', encoding='utf-8', index=False)
```
